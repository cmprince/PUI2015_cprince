```python
#to store the data i am generating an empty array of the size of the values 
#of n i want to test. i do that with np.zeros which takes the length of the 
#array as argument, and the data type as optional argument, but default is 
#float, which is fine by me, so i do not need any argument other then the lenght 
narray = range(1,50,1)
parray = np.logspace(-2, np.log10(0.5), 10)
#parray = np.linspace(0.05, 0.5, 6)

#print parray
ks_b = np.zeros((len(narray), len(parray)))
ad_b = np.zeros((len(narray), len(parray)))
kl_b = np.zeros((len(narray), len(parray)))
chi2_b = np.zeros((len(narray), len(parray)))


def mynorm (x, mu, var):
    return scipy.stats.norm.cdf(x, loc=mu, scale=var)

#then i put the tests in a for loop so that i can generate a distribution for given parameters once, 
#and run all tests against it
#now the valus that i want to plot depends on how i intend to describe the plot, and viceversa. 

#here is one way to plot it, knowing the values i get for the tests when 
#i throw in a poisson distribution with low l and compare it with a gaussian and 
#assuming that that is a vary bad match
p=0.5
for i,n in enumerate(narray):
    for j,p in enumerate(parray):
        #p=0.1 #parameter for the binomial, my arbitrary choice
        #generate the distribution
        dist = np.random.binomial(n, p, 1000)
        #run the tests. 

        ks_b[i][j] = scipy.stats.kstest(dist, mynorm, args=(n*p, n*p*(1.0-p)))[0]
        ad_b[i][j] = scipy.stats.anderson(dist, dist='norm')[0]


        # for KL and Pearson's chisq I have to simulate the normal distribution as well
        mybins=np.linspace(min(dist),max(dist), 25) 
        bincenters = mybins[:-1]+0.5*(mybins[1]-mybins[0])

        #when i was coding this up something was wrong. i put some plots in to figure out what... just so you know.
        #if i%10 == 0: 
        #    print n
        #    pl.figure()
        #pl.hist(dist, bins=mybins)
        #pl.plot(bincenters, scipy.stats.norm.pdf(bincenters, loc=n*p, scale=n*p*(1-p)))
        #print np.histogram(dist, bins=mybins)[0]
        #print scipy.stats.norm.pdf(bincenters, loc=n*p, scale=n*p*(1.0-p))
        kl_b[i][j] =  scipy.stats.entropy(np.histogram(dist, bins=mybins)[0], 
                                          scipy.stats.norm.pdf(bincenters, loc=n*p, scale=n*p*(1.0-p)))
        chi2_b[i][j] = scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0], 
                                          scipy.stats.norm.pdf(bincenters, loc=n*p, scale=n*p*(1.0-p)))[0]
```
