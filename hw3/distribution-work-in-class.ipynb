```python
from __future__ import print_function
import pylab as pl
import numpy as np

%pylab inline
```

```python
distribs = ['normal','poisson','binomial','chisquare','weibull','']
```

```python
mydist = np.random.chisquare(df=100,size=100)
```

```python
pl.hist(mydist, bins=25)
```

```python
mymean = 100
#sizes = range(10,2000,100)
distributions = {}
distributions['chisq'] = {}
```

```python
sizes = 1990/(np.array(linspace(1,1990,100))+10)
```

```python
for n in sizes:
    distributions['chisq'][n] = np.random.chisquare(mymean, n)
```

```python
fig_mu_N = pl.figure(figsize=(10,10))
axchisq_mu_N = fig_mu_N.add_subplot(111)
for nn in distributions['chisq'].iterkeys():
    axchisq_mu_N.plot(nn, distributions['chisq'][nn].mean(), 'o')
    axchisq_mu_N.set_xlabel('sample size')
    axchisq_mu_N.set_ylabel('sample mean')
    axchisq_mu_N.plot([min(sizes), max(sizes)], [mymean,mymean], 'k')
```
